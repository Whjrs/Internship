# Q1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views

```python
driver = uc.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
time.sleep(2)

Rank=[]
Name=[]
Artist=[]
UploadDate=[]
Views=[]

rank = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
try:
    for i in rank[0:30]:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("-")


name = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
try:
    for i in name[0:30]:
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-")

artist = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
try:
    for i in artist[0:30]:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")

upload = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
try:
    for i in upload[0:30]:
        UploadDate.append(i.text)
except NoSuchElementException:
    UploadDate.append("-")
    
views = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
try:
    for i in views[0:30]:
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")

print("Rank=",len(Rank))
print("Name=",len(Name))
print("Artist=",len(Artist))
print("UploadDate=",len(UploadDate))
print("Views=",len(Views))

df = pd.DataFrame({'Rank': Rank, 'Name': Name,'Artist': Artist, 'Upload Date': UploadDate,'Views': Views})
df
```
