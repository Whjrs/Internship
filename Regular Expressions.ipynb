# Write a Python program to check that a string contains only a certain set of characters (in this case a-z, A-Z and 0-9).

```python

EnteredString = input('Enter a string: ')
Pattern = re.compile('[^a-zA-Z0-9]')
String = Pattern.search(EnteredString)
if String:
    print('The string contains',String.group())
else:
    print('The string contains a-z,A-Z and 0-9')
    
```

# Create a function in python that matches a string that has an a followed by zero or more b's

```python

EnteredString = input('Enter a string: ')
def match(String):
    Pattern = re.compile('ab*?')
    String = Pattern.search(EnteredString)
    if re.search(Pattern, EnteredString):
        return 'Found a match'
    else:
        return 'Not a match'
print(match(EnteredString))
```

# Write a Python program to search some literals strings in a string

# Write a Python program to search a literals string in a string and also find the location within the original string where the pattern occurs

# Write a Python program to find the substrings within a string.

# Write a Python program to find the occurrence and position of the substrings within a string.

```python
String = 'Python exercises, PHP exercises, C# exercises'
Substring='exercises'
LiteralString = re.finditer(Substring, String)
for i in LiteralString:
    Start = i.start()
    End = i.end()
    print("The substring ",Substring,"found at position", Start,":",End)
```

# Write a Python program to convert a date of yyyy-mm-dd format to dd-mm-yyyy format.

```python
Date = '2018-03-31'
Pattern = re.split('-', Date)
NewDate = '-'.join(Pattern[::-1])
print(NewDate)
```

# Write a Python program to separate and print the numbers and their position of a given string.

```python
EnteredString = input('Enter a string: ')
for i in re.finditer('\d+',EnteredString):
    print("The number in the string is: ",i.group(0))
    print("Index position:", i.start())
```

# Write a Python program to remove duplicate words from Sentence using Regular Expression

```python
EnteredString = input('Enter a string: ')
String = re.sub("\\b(\\w+)(?:\\W+\\1\\b)+",r'\1',EnteredString)
print(String)
```
