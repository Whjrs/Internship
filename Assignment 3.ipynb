# Q1. Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.                                                                                                                                                #Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”.
In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
driver = webdriver.Chrome()
driver.get("https://www.amazon.in/")

Product_input = driver.find_element(By.ID,"twotabsearchtextbox")
product_name=input("Enter the Product name you want to search for : ")
Product_input.send_keys(product_name)
Submit_btn = driver.find_element(By.XPATH,"//input[@type='submit']")
Submit_btn.click()

Product_URL = []
URL = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')

for i in URL:
    Product_URL.append(i.get_attribute("href"))
len(Product_URL)

BrandName =[]
NameOfProduct=[]
Price=[]
Return=[]
ExpectedDelivery=[]
Availability=[]

for i in Product_URL:
    driver.get(i)
    time.sleep(6)
    try:
        Brand = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[46]/div/table/tbody/tr[1]/td[2]/span')
        BrandName.append(Brand.text)
    except NoSuchElementException:
        BrandName.append('-')
    
    try:
        name = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div/h1/span')
        NameOfProduct.append(name.text)
    except NoSuchElementException:
        NameOfProduct.append('-')
        
    try:
        price = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[2]/div[1]/div/span[1]/span[2]/span[2]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
        
    try:
        ret = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span')
        Return.append(ret.text)
    except NoSuchElementException:
        Return.append('-')
        
    try:
        Exp = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[2]/div[9]/div[1]/div/div/div/span/span')
        ExpectedDelivery.append(Exp.text)
    except NoSuchElementException:
        ExpectedDelivery.append('-')
        
    try:
        avail = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[4]/div/div[1]/span')
        Availability.append(avail.text)
    except NoSuchElementException:
        Availability.append('-')

df = pd.DataFrame({'Brand Name': BrandName, 'Name Of Product': NameOfProduct,'Price': Price, 'Return/Replacement': Return,'Expected Delivery': ExpectedDelivery, 'Availability ': Availability, 'Product URL':Product_URL})
df
    
```

# Q4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV. 

```python
import re
driver = webdriver.Chrome()
driver.get("https://www.flipkart.com/")

time.sleep(5)

try:
    driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _2doB4z"]').click()
except:
    print("No pop-up-1")

Product_input = driver.find_element(By.CLASS_NAME,"_3704LK")
product_name=input("Enter the Product name you want to search for : ")
Product_input.send_keys(product_name)

Search = driver.find_element(By.XPATH,'//button[@class="L0Z3Pu"]')
Search.click()




BrandName =[]
SmartPhoneName=[]
Color=[]
RAM=[]
Storage=[]
PrimaryCamera=[]
SecondaryCamera=[]
DisplaySize=[]
BatteryCapacity=[]
Price=[]
Product_URL = []


time.sleep(6)

Brand = driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in Brand:
    try:
        BrandName.append(i.text[0:12])
    except NoSuchElementException:
        BrandName.append('-')
    
name = driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in name:
    try:
        SmartPhoneName.append(i.text)
    except NoSuchElementException:
        SmartPhoneName.append('-')
        
price = driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
for i in price:
    try:
        Price.append(i.text)
    except NoSuchElementException:
        Price.append('-')
        
col = driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in col:
    try:
        string=re.findall(r'\((.*?)\,',i.text)
        Color.append(string[0])
    except NoSuchElementException:
        Color.append('-')
        
ram = driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
for i in ram:
    try:
        RAM.append(i.text[0:9])
    except NoSuchElementException:
        RAM.append('-')
        
avail = driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
for i in avail:
    try:
        Storage.append(i.text[11:17])
    except NoSuchElementException:
        Storage.append('-')
    
PC = driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
for i in PC:
    try:
        string = i.text.splitlines()[2]
        PrimaryCamera.append(string.split("|"))
        SecondaryCamera.append(string.split("|")[0])
    except NoSuchElementException:
        PrimaryCamera.append('-')
        SecondaryCamera.append('-')

    
DS = driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
for i in DS:
    try:
        DisplaySize.append(i.text.splitlines()[1])
    except NoSuchElementException:
        DisplaySize.append('-')
        
BC = driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
for i in BC:
    try:
        BatteryCapacity.append(i.text.splitlines()[3])
    except NoSuchElementException:
        BatteryCapacity.append('-')
        
URL = driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in URL:
    try:
        Product_URL.append(i.get_attribute("href"))
    except NoSuchElementException:
        Product_URL.append('-')
    
    

df = pd.DataFrame({'Brand Name': BrandName, 'Smart Phone Name': SmartPhoneName,'Price': Price, 'Color': Color,'RAM': RAM, 'Storage ': Storage,'Primary Camera': PrimaryCamera,'Secondary Camera': SecondaryCamera,'Display Size': DisplaySize, 'Battery Capacity': BatteryCapacity,'Product_URL ': Product_URL})
df
```

# Q5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
import re
driver = webdriver.Chrome()
driver.get("https://www.google.com/maps/")

UserInput=input("Enter the City name : ")
Search=driver.find_element(By.XPATH,'//input[@id="searchboxinput"]')
Search.send_keys(UserInput)

Search = driver.find_element(By.XPATH,'//button[@id="searchbox-searchbutton"]')
Search.click()

time.sleep(4)

Lat = []
Long = []
try:
    Current_URL = driver.current_url
    Lat_Long = re.findall(r'@(.*)data',Current_URL)
    if len(Lat_Long):
        lat_long = Lat_Long[0].split(",")
        if len(lat_long)>=2:
            Lat.append(lat_long[0])
            Long.append(lat_long[1])
except Exception as e:
    print("Error : ", str(e))

time.sleep(2)

df = pd.DataFrame({'City': UserInput, 'Latitude': Lat, 'Longitude': Long})
df
```

# Q8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video. 

```python
driver = webdriver.Chrome()
URL=input("Enter the URL: ")
driver.get(URL)


Comments = []
Upvote = []
Times = []

for _ in range(30):
    driver.execute_script("window.scrollBy(0,100)")
    time.sleep(5)
    comment = driver.find_elements(By.ID,"content-text")
    for i in comment:
        try:
            Comments.append(i.text)
        except:
            Comments.append("-")
        
    upvote = driver.find_elements(By.ID,"vote-count-middle")
    for i in upvote:
        try:
            Upvote.append(i.text)
        except:
            Upvote.append("-")
        
    times = driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
    for i in times:
        try:
            Times.append(i.text)
        except:
            Times.append("-")

print(len(Comments))  
print(len(Upvote))
print(len(Times))

df=pd.DataFrame({'Comments':Comments,'Comments Upvote':Upvote,'Time when comment is posted':Times})
df
```
