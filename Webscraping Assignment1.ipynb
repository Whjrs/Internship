# Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#request the webpage server to get the source code
page= requests.get('https://en.wikipedia.org/wiki/Main_Page')
page

#Scraping the page content
soup = BeautifulSoup(page.content)

#Scraping all the header tags by looping
heading_tag = ['h1','h2','h3','h4','h5','h6']
header=[]
for i in soup.find_all(heading_tag):
    header.append(i.text)
#display(header)

# making Dataframe
df = pd.DataFrame({"Header " : header})
df
```

# Write a python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
page= requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page

#Scraping the page content
soup = BeautifulSoup(page.content)

#Scraping all the name by looping
heading_tag = ['h3']
h3_tag=[]
names=[]
for i in soup.find_all(heading_tag):
    h3_tag.append(i.text)
    names = [j.split('(',1)[0] for j in h3_tag]
#display(names)

#Scraping term of office

term_of_office=[]
for i in soup.find_all('p', class_=""):
    term_of_office.append(i.text.split(":")[1])
for i in sorted([1,3,5,7],reverse=True):
    del term_of_office[i]
#display(term_of_office)

# making Dataframe
df = pd.DataFrame({"Names " : names, "Term of office ": term_of_office})
df
```

# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating

```python
#a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
page= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
#Scraping the page content
soup = BeautifulSoup(page.content)

team=[]
for i in soup.find_all('span', class_="u-hide-phablet"):
    team.append(i.text)
#display(team[0:10])

matches=[]
for i in soup.find_all('td', class_="rankings-block__banner--matches"):
    matches.append(i.text)
for i in soup.find_all('tr',class_="table-body"):
    matches.append(i.text.replace('\n'," ").replace('NZ',"").replace('SA',"").replace('WI',"").replace('SL',"").strip().split()[3])
#display(matches[0:10])

points=[]
for i in soup.find_all('td', class_="rankings-block__banner--points"):
    points.append(i.text)
for i in soup.find_all('tr',class_="table-body"):
    points.append(i.text.replace('\n'," ").replace('NZ',"").replace('SA',"").replace('WI',"").replace('SL',"").strip().split()[4])
#display(points[0:10])

ratings=[]
for i in soup.find_all('td', class_="rankings-block__banner--rating u-text-right"):
    ratings.append(i.text.strip())
for i in soup.find_all('tr',class_="table-body"):
    ratings.append(i.text.replace('\n'," ").replace('NZ',"").replace('SA',"").replace('WI',"").replace('SL',"").strip().split()[5])
#display(ratings[0:10])

df = pd.DataFrame({"Team " : team[0:10], "Matches ": matches[0:10],"Points ": points[0:10], "Ratings ": ratings[0:10]})
df



```

```python
#b) Top 10 ODI Batsmen along with the records of their team and rating.
page= requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
page
#Scraping the page content
soup = BeautifulSoup(page.content)

bRating=[]
for i in soup.find('div',class_="rankings-block__banner--rating"):
    bRating.append(i.text)

for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    bRating.append(i.text)

bTeam=[]
bTeam.append('PAK')
for i in soup.find_all('span',class_="table-body__logo-text"):
    bTeam.append(i.text)

df=pd.DataFrame.from_dict({'Team':bTeam[0:10],'Rating':bRating[0:10]},orient='index')
odiBatsMen=df.transpose()
odiBatsMen
```

```python
#c) Top 10 ODI bowlers along with the records of their team and rating
page= requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
page
#Scraping the page content
soup = BeautifulSoup(page.content)

BowlRating=[]
for i in soup.find('div',class_="rankings-block__banner--rating"):
    BowlRating.append(i.text)

for i in soup.find_all('td',class_="table-body__cell rating"):
    BowlRating.append(i.text)

BowlTeam=[]
for i in soup.find_all('div',class_="rankings-block__banner--nationality"):
    BowlTeam.append(i.text.strip())
        
for i in soup.find_all('span',class_="table-body__logo-text"):
    BowlTeam.append(i.text)
    
df=pd.DataFrame.from_dict({'Team':BowlTeam[0:10],'Rating':BowlRating[0:10]},orient='index')
odiBowlMen=df.transpose()
odiBowlMen
```

# Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame
i) Headline
ii) Time
iii) News Link

```python

page= requests.get('https://www.cnbc.com/world/?region=world')
page
#Scraping the page content
soup = BeautifulSoup(page.content)

# Headline
Headline=[]
for i in soup.find_all('div',class_="LatestNews-headlineWrapper"):
    Headline.append(i.text.split("Ago")[1])
#display(Headline)

# Time   
Time=[]    
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    Time.append(i.text)
#display(Time)

# NewsLink
NewsLink=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    NewsLink.append(i["href"])
#display(NewsLink)

df=pd.DataFrame({"Headline ": Headline, "Time ": Time ,"NewsLink ": NewsLink})
df
```

# Write a python program to scrape the details of most downloaded articles from AI in last 90days.
https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details and make data frame
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL


```python
page= requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page
#Scraping the page content
soup = BeautifulSoup(page.content)


#Paper Title
PaperTitle=[]
for i in soup.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    PaperTitle.append(i.text)
#display(PaperTitle)

#Authors
Authors=[]
for i in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    Authors.append(i.text)
#display(Authors)

#Published Date
PublishedDate=[]
for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    PublishedDate.append(i.text)
#display(PublishedDate)

#Paper URL
PaperURL=[]
for i in soup.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    PaperURL.append(i['href'])
#display(PaperURL)

df=pd.DataFrame({"Paper Title ": PaperTitle, "Authors ": Authors ,"Published Date ": PublishedDate,"Paper URL ": PaperURL})
df
```

# Write a python program to scrape mentioned details from dineout.co.inand make data frame
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image UR

```python
page= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
#Scraping the page content
soup = BeautifulSoup(page.content)

#Restaurant name
Restaurantname=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    Restaurantname.append(i.text)
#display(Restaurantname)

#Cuisine
Cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    Cuisine.append(i.text.split("|")[1])
#display(Cuisine)

#Location
Location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
#display(Location)

#Ratings
Ratings=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    Ratings.append(i.text)
#display(Ratings)

#Image URL
ImageURL=[]
for i in soup.find_all('img',class_="no-img"):
    ImageURL.append(i['data-src'])
#display(ImageURL)


df=pd.DataFrame({"Restaurant Name ": Restaurantname, "Cuisine ": Cuisine ,"Location ": Location,"Ratings ": Ratings,"ImageURL ":ImageURL})
df   
```
