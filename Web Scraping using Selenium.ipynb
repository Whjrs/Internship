# Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and salary filter. You have to scrape data for “Data Scientist” designation for first 10 job results. You have to scrape the job-title, job-location, company name, experience required. 

```python
driver = webdriver.Chrome()
driver.get("http://www.naukri.com/")

Designation = driver.find_element(By.CLASS_NAME,"suggestor-input")
Designation.send_keys('Data Scientist')

Location = driver.find_element(By.XPATH,'//input[@placeholder="Enter location"]')
Location.send_keys('Delhi/NCR')

Search = driver.find_element(By.CLASS_NAME,"qsbSubmit")
Search.click()

time.sleep(3)

job_title=[]
job_location=[]
company_name=[]
experience_required=[]

title_tags = driver.find_elements(By.XPATH,'//a[@class="title "]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)


location_tags = driver.find_elements(By.XPATH,'//span[@class="locWdth"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)

company_tags = driver.find_elements(By.XPATH,'//a[@class=" comp-name mw-25"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)

experience_tags = driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in experience_tags[0:10]:
    experience=i.text
    experience_required.append(experience)
    
    
df = pd.DataFrame({'Job Title': job_title, 'Job Location': job_location,'Company Name': company_name, 'Experience required': experience_required})
df
```

As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100reviews.


```python
driver = webdriver.Chrome()
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market")

time.sleep(5)

try:
    driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _2doB4z"]').click()
except:
    print("No pop-up")

time.sleep(3)

rating=[]
review_summary=[]
full_review=[]

for page in range(0,10):
    rating_tags = driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
    for i in rating_tags[0:100]:
        rate=i.text
        rating.append(rate)
    review_tags = driver.find_elements(By.XPATH,'//p[@class="z9E0IG"]')
    for i in review_tags[0:100]:
        review=i.text
        review_summary.append(review)
        
    fullreview_tags = driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]')
    for i in fullreview_tags[0:100]:
        freview=i.text
        full_review.append(freview)
        
time.sleep(3)
nextpage = driver.find_element(By.XPATH,'//a[@class="_9QVEpD"]')
nextpage.click()


df = pd.DataFrame({'Rating': rating[0:100], 'Review Summary': review_summary[0:100],'Full Review': full_review[0:100]})
df
```

# Q5: Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then set CPU
Type filter to “Intel Core i7” Aftersetting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. Title
2. Ratings
3. Price

```python
driver = webdriver.Chrome()
driver.get("https://www.amazon.in/")

time.sleep(2)

Designation = driver.find_element(By.XPATH,'//input[@class="nav-input nav-progressive-attribute"]')
Designation.send_keys('Laptop')

time.sleep(2)

Search = driver.find_element(By.XPATH,'//span[@class="nav-search-submit-text nav-sprite nav-progressive-attribute"]')
Search.click()


time.sleep(2)

title=[]
rating=[]
price=[]

title_tags = driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
for i in title_tags[0:10]:
    b=i.text
    title.append(b)

rating_tags = driver.find_elements(By.XPATH,'//i//span[@class="a-icon-alt"]')
for i in rating_tags:
    p=i.get_attribute("innerHTML")
    rating.append(p)
    
price_tags = driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
for i in price_tags[0:10]:
    pr=i.text
    price.append(pr)


df = pd.DataFrame({'Title': title[0:10], 'Rating': rating[0:10],'Price': price[0:10]})
df
```

# Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuote
3. Than scrap a)Quote b) Author c) Type Of Quote

```python
driver = webdriver.Chrome()
driver.get("https://www.azquotes.com/")

time.sleep(2)

driver.find_element(By.XPATH,'//a[@href="/top_quotes.html"]').click()
time.sleep(2)

quote=[]
author=[]
type_quote=[]

for page in range(0,10):
    
    quote_tags = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tags[0:1000]:
        b=i.text
        quote.append(b)
    
    author_tags = driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tags[0:1000]:
        p=i.text
        author.append(p)
        
    
    type_quote_tags = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_quote_tags[0:1000]:
        pr=i.text
        type_quote.append(pr)
        
time.sleep(3)
nextpage = driver.find_element(By.XPATH,'//li[@class="next"]')
nextpage.click()

df = pd.DataFrame({'Quote': quote, 'Author': author,'Type Quote': type_quote})
df
```

# Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name,
Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1
scrap the mentioned data and make the DataFrame

```python
driver = webdriver.Chrome()
driver.get("https://www.jagranjosh.com/general-knowledge/list-of%02all-prime-ministers-of-india-1473165149-1")

n=[]
Name=[]
Age=[]
Term=[]
term=[]
Duration=[]

xpath='//article[@id="itemdiv"]/div[4]/div[5]/div/table/tbody/tr['

for i in range(2,20):
    path=xpath+str(i)+"]"
    for j in range(2,7):
        finalpath=path+"/td["+str(j)+"]"
        name_tag = driver.find_elements(By.XPATH,finalpath)
        for k in name_tag:
            n.append(k.text)
    
    Name.append(n[0])
    Age.append(n[1])
    Term.append(n[2])
    term.append(n[3])
    Duration.append(n[4])
    n=[]
            
df = pd.DataFrame({'Name': Name, 'Age when assumed office': Age,'Term of office from': Term,'Term of office to': term,'Duration':Duration})
df
```
